Import the necessery modules

```python
import pandas as pd
```

Import both the strava and the strong data in a dataframe so we can take a closer look

```python
dfstrava_raw = pd.read_csv('strava_data.txt')
dfstrava_raw.head()
```

```python
dfstrong_raw = pd.read_csv('strong.csv')
dfstrong_raw.tail(60)
```

We're going to only use the columns we want use from the strava file and name those columns so that they match the ones in the strong file

```python
dfstrava = pd.read_csv('strava_data.txt', header=0, usecols=['start_date', 'name', 'distance', 'moving_time', 'type'])
dfstrava.columns =['Date', 'Workout Name', 'Distance', 'Seconds', 'Exercise Name']
dfstrava
```

Changing the exercise with name from "Ride" to "Cycling" so that they match the strong name

```python
dfstrava = dfstrava.replace('Ride', 'Cycling')
```

Changing the distances in the strong data file to metres from km, so we can compare apples to apples

```python
dfstrong_raw.Distance = dfstrong_raw.Distance*1000
dfstrong_raw.tail(70)
```

Adding the two dataframes into one

```python
dfcombo = dfstrong_raw.append(dfstrava, 'sort=False')
dfcombo
```
