Import the necessary modules

```python
import pandas as pd
```

Loading in the dataset

```python
df = pd.read_csv('strong.csv')
df.head()
```

Adding an column that calculates the weigt lifted per set

```python
df['Lifted'] = (df['Weight'] * df['Reps'] ) 
df.head()
```

Making a pivot table that shows the total amount lifted per exercis

```python
pivot = df.pivot_table(index = ['Exercise Name'], values = ['Lifted'], aggfunc = 'sum')
pivot
```
