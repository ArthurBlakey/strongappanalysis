# Import the necessery modules

```python
import pandas as pd
```

# Taking a glance at the datasets
Import both the strava and the strong data in a dataframe so we can take a closer look

```python
dfstrava= pd.read_csv('activities.csv')
dfstrava.head()
```

```python
dfstrong = pd.read_csv('strong.csv')
dfstrong.head()
```

# Manipulating the dataframes

Adding a column with the source

```python
dfstrong['Source'] = 'strong'
dfstrava['Source'] = 'strava'
```

We're going to only use the columns we want use from the strava file and name those columns so that they match the ones in the strong file

```python
dfstrava = pd.read_csv('activities.csv', header=0, usecols=['When', 'Type', 'Dist km', 'Elapsed Time', 'Speed km/h'])
dfstrava.columns =['Date', 'Exercise Name','Distance', 'Seconds', 'Speed']
dfstrava.head()
```

Changing the exercise with name from "Ride" to "Cycling" so that they match the strong name

```python
dfstrava = dfstrava.replace('Ride', 'Cycling')
```

Converting the distance in the strong file from meter to km, to allign with strava data

```python
dfstrong.Distance = dfstrong.Distance*1000
```

Computing the value for 'Speed' in strong dataframe

```python
dfstrong['Speed'] = dfstrong["Distance"] / (dfstrong["Seconds"]/3.6)
```

Coverting the value for speed in the strava file from m/s to km/u

```python
dfstrava['Speed'] = dfstrava["Distance"] / (dfstrava["Seconds"]/3.6)
```

<h1>Adding the two dataframes into one

```python
dfcombo = dfstrong.append(dfstrava)
dfcombo = dfcombo.sort_values('Date')
dfcombo
```
